### Data Exploration Analysis

```python
# a frequency plot to show the number of null values in each column of the DataFrame
df.isnull().sum().plot(kind='bar', figsize=(10,5))
plt.title('Null values per column')
plt.show()
```

```python
# Missing Value Treatment
## Remove the records whose Closed Date values are null
df=df[df['Closed Date'].notnull()]
```

```python
df.isnull().sum()
```

```python
df['Created Date'] = pd.to_datetime(df['Created Date'], errors='coerce')
```

```python
df['Closed Date'] = pd.to_datetime(df['Closed Date'], errors='coerce')
```

```python
df.loc[:,'Time_Elapsed'] = df['Closed Date'] - df['Created Date']
```

```python
df[['Created Date', 'Closed Date']].dtypes
```

```python
df.loc[:,'Request_Closing_Time'] = (df['Closed Date'] - df['Created Date']).dt.total_seconds()
```

```python
# frequency plot for complaints in each city
df['City'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title("complaints by city")
plt.show()
```

```python
# a scatter and hexbin plot of the concentration of complaints across Brooklyn
brooklyn = df[df['City'] == 'BROOKLYN']
plt.scatter(brooklyn['Longitude'], brooklyn['Latitude'])
plt.title("Complaints across brooklyn")
plt.show()
brooklyn.plot.hexbin(x='Longitude', y='Latitude', gridsize=20)
plt.show()
```

```python
# bar graph to show the types of complaints
df['Complaint Type'].value_counts().plot(kind='bar', figsize=(10,5))
plt.show()
```

```python
# frequency of various types of complaints for New York City
nycomplaint_df = df[df['City'] == 'NEW YORK']    # data filter# 
complaint_counts = nycomplaint_df['Complaint Type'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
complaint_counts.plot(kind='bar')
plt.title("Frequency of complaint types in new york city")
plt.xlabel("Complaint Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 complaint

top10 = complaint_counts.head(10)
plt.figure(figsize=(10,5))
top10.plot(kind='bar')
plt.title("Top 10 Complaint in new york city")
plt.xticks(rotation=45)
plt.show()
```

```python
# various types of complaint in each city

df_new = pd.crosstab(df['Complaint Type'], df['City'])
#df_new
df_new.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Complaint types in Cities")
plt.show()
```

```python
# Sort the complaint types based on the average Request_Closing_Timegrouping them for different locations
avg_df=df.groupby(['City','Complaint Type'])['Request_Closing_Time'].mean()

avg_df=avg_df.reset_index()   # conversion to dataframe for clear

sort_df=avg_df.sort_values(by='Request_Closing_Time', ascending=False)    # not on series , useful on dataframe
sort_df.head(5)
```

```python
# See whether the average response time across different complaint types is similar (overall)
response_time_avg=df.groupby('Complaint Type')['Request_Closing_Time'].mean()

response_time_avg.plot(kind='bar', figsize=(10,8))
plt.title('Average response time across complaint types', color='red')
plt.xticks(rotation=45)
plt.ylabel('Avg time')
plt.show()
```

```python
# kruskel wallis h test
from scipy.stats import kruskal

groups = [group['Request_Closing_Time'] for _, group in df.groupby('Complaint Type')]

stat, p = kruskal(*groups)

print(p)
```

```python
groups_city = [group['Request_Closing_Time'] for _, group in df.groupby('City')]

stat, p_city = kruskal(*groups_city)

print("P-value (City):", p_city)
```

##### Observation after performing the kruskel wallisH test for significan variables
-- The p-value we get for complaint type and city is 0.0 which is less than 0.05 means it reject null hypothesis
-- p < 0.05 reject null hypothesis(Ho)
